### Functions

$$
F: D \to R
$$

- **D** is the **domain** – the set of inputs.
- **R** is the **Range** or **codomain**  – the set of possible outputs.
- **F** is a **function** – a rule that assigns each element in **D** exactly one value in **R**.

## Example f

$$
\begin{aligned}
D &= \mathbb{R} \\
R &= \mathbb{R} \\
f &: D \to R \\
f(x) &= x^2 \\

\mathbb{R} &:\ \text{the set of all real numbers}

\end{aligned}
$$




## Example g

$$
\begin{aligned}
D &= \{ x \in \mathbb{R} \mid x > 0 \} \\
R &= \mathbb{R} \\
g &: D \to R \\
g(x) &= x^2 \\
\\
\end{aligned}
$$

* g(x) is not the same as f(x)!
*  Eventhough the Rule is the same (y=x^2), they differ in their Range!
* Function are defined by their 3 Components: Domain, Range, Rule

## What is the Graph of f(x)? 

## The Graph of a Function

The **graph** of a function $f: D \to R$ is a set of ordered pairs defined by the rule $f(x)$. It is a **subset** of the Cartesian product $D \times R $:

$$
\text{Graph}(f) = \{\, (x, f(x)) \mid x \in D \,\} \subseteq D \times R
$$

Each point on the graph represents one input–output pair: the input \( x \) from the domain \( D \), and its corresponding output \( f(x) \) in the codomain \( R \).

```python
import numpy as np
from matplotlib import pyplot as plt
from itertools import product

def f(x: float) -> float:
    return 4 * x**2 + 12 * x - 24

# Sample domain
X = np.linspace(-20, 20, 500)  # smaller range for plotting
Y = f(X)

# Cartesian product: X × Y (not X × X)
cartesian_prod = np.array(list(product(X, Y)))


# Only keep (a, b) where f(a) ≈ b (within tolerance)
plot_points = [(a, b) for a, b in cartesian_prod if np.isclose(f(a), b, atol=1e-5)]

print(f"Found {len(plot_points)} points that match f(x).")

# Plot function curve
plt.plot(X, Y, label="f(x)", color="blue")

# Plot matching grid points (they should all lie on the curve)
if plot_points:
    pts_x, pts_y = zip(*plot_points)
    plt.scatter(pts_x, pts_y, color='red', label="Matching points", s=10)


plt.legend()
plt.grid(True)
plt.title("f(x) = 4x² + 12x - 24")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show() 
```
